## Regresia

V analytickej chémií často využívame lineárnu regresiu. V nasledujúcej úlohe si vyskúšame, ako tento problém riešiť pomocou pythonu.

Na túto úlohu budeme používať knižnicu **numpy**.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# vytvorime si ukazkove data

c = list(np.arange(0, 10, 0.5))
sig = [10*i + random.uniform(0,10) for i in c]
```

```python
# vypocitame parametre regresie

fit = np.polyfit(c, sig, 1)
```

```python
# vykreslime do grafu data aj regresiu

plt.plot(c,sig, linestyle='', marker='o')
plt.plot(c, [fit[0] * x + fit[1] for x in c], color='red')
plt.title('regrese')
plt.xlabel('koncentrace')
plt.ylabel('signal')
plt.annotate('y = {:.2}x+{:.2}'.format(fit[0],fit[1]),xy=(6,10))
```

## Exploratórna analýza dát

```python
import scipy 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris
from pandas.plotting import scatter_matrix
from pandas.plotting import parallel_coordinates
from sklearn.decomposition import PCA
```

```python
# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, names=names)
```

```python
# descriptions
print(dataset.describe())
```

```python
# class distribution
print(dataset.groupby('class').size())
print(dataset.groupby('class').describe())
```

```python
# box and whisker plots
dataset.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
sns.boxplot(ax = ax1, x="class", y="sepal-length", data=dataset)
sns.boxplot(ax = ax2, x="class", y="sepal-width", data=dataset)
sns.boxplot(ax = ax3, x="class", y="petal-length", data=dataset)
sns.boxplot(ax = ax4, x="class", y="petal-width", data=dataset)
plt.show()
```

```python
# histograms
dataset.hist()
plt.show()
```

```python
# scatter plot matrix
scatter_matrix(dataset)
plt.show()
```

```python
#parallel coordinates
parallel_coordinates(dataset, 'class')
plt.show()
```

```python
#pair plot
sns.pairplot(dataset, hue='class', height=2.5)
plt.show()
```

```python
# violin plot
sns.violinplot("class", "sepal-length", data=dataset)
plt.show()
```

## Analýza hlavných komponent

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(dataset.iloc[:,:4])
df = pd.DataFrame(reduced_data, columns=list('xy'))
df['class'] = dataset['class']
```

```python
groups = df.groupby('class')
```

```python
fig, ax = plt.subplots()
ax.margins(0.05) 
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=12, label=name)
ax.legend()
fig.suptitle('PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
```
